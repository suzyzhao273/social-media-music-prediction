```python
cols = [
    "music_volume_baseline",
    "music_volume_release",
    "music_controversy_baseline",
    "music_controversy_release"
]


summary = pd.DataFrame({
    "non_missing": df[cols].notna().sum(),
    "zero_count": (df[cols].fillna(0) == 0).sum(),
    "nonzero_count": (df[cols].fillna(0) > 0).sum(),
    "nonzero_share": (df[cols].fillna(0) > 0).mean()
})

summary
    
```

```python
release_check = pd.DataFrame({
    "metric": ["volume_release", "controversy_release"],
    "nonzero_count": [
        (df["music_volume_release"].fillna(0) > 0).sum(),
        (df["music_controversy_release"].fillna(0) > 0).sum()
    ],
    "nonzero_share": [
        (df["music_volume_release"].fillna(0) > 0).mean(),
        (df["music_controversy_release"].fillna(0) > 0).mean()
    ]
})

release_check
```
